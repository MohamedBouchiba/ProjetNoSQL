```python
import os
import sys
import json
import pandas as pd
```

```python
# moviesfile = open("data/movies.dat", "r")
# ratingsfile = open("data/ratings.dat", "r")
# usersfile = open("data/users.dat", "r")

list_movies=[]
list_ratings=[]
list_users=[]
```

```python
with open("data/movies.dat", "r") as moviesfile, open("data/ratings.dat", "r") as ratingsfile, open("data/users.dat", "r") as usersfile:
    
    for elem in moviesfile.readlines():
        movie={}
        elem = elem.replace("\n","")
        contenue = elem.split("::")    
        movie["movie_id"]= int(contenue[0])
        movie["title"]= contenue[1]
        movie["genres"]= contenue[2].split("|")
        list_movies.append(movie)
    
    for elem in ratingsfile.readlines():
        rating= {}
        elem = elem.replace("\n","")
        contenue = elem.split("::")
        rating["user_id"]=int(contenue[0])
        rating["movie_id"]=int(contenue[1])
        rating["rating"]=int(contenue[2])
        rating["timestamp"]=int(contenue[3])
        list_ratings.append(rating)

    for elem in usersfile.readlines():
        user= {}
        elem = elem.replace("\n","")
        contenue = elem.split("::")
        user["user_id"]=int(contenue[0])
        user["genres"]= contenue[1]
        user["age"]=int(contenue[2])
        user["occupation"]=int(contenue[3])
        user["zip"]=contenue[4]
        list_users.append(user)
```

```python
df_ratings = pd.DataFrame(list_ratings)
df_ratings.head(5)
```

```python
df_users= pd.DataFrame(list_users)
df_users.head(5)
```

```python
df_movies = pd.DataFrame(list_movies)
df_movies.head(5)
```

```python
tmp = pd.merge(df_ratings, df_users, on='user_id')
tmp.head(5)
```

```python
data = pd.merge(tmp, df_movies, on="movie_id")
data.head(10)
```
